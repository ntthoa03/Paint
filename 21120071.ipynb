## Câu 2:
Explain the significance of Term Frequency (TF) and Inverse Document Frequency (IDF) in text analysis. Why is it essential to use both TF and IDF together?

Term Frequency (TF) và Inverse Document Frequency (IDF) là hai khái niệm quan trọng trong phân tích văn bản (text analysis) . Chúng đóng vai trò quan trọng trong việc đánh giá mức độ quan trọng của một từ hoặc cụm từ trong tài liệu văn bản.

- **Term Frequency (TF):**
  - TF đo lường tần số xuất hiện của một từ hoặc cụm từ trong một văn bản cụ thể, được tính bằng cách đếm số lần xuất hiện của từ đó trong văn bản và sau đó chia cho tổng số từ trong văn bản đó.
  - TF là một cách đơn giản và dễ hiểu để đánh giá mức độ quan trọng của một từ trong văn bản. Một từ xuất hiện càng nhiều lần trong một tài liệu thì nó càng quan trọng để mô tả nội dung của tài liệu đó.Tuy nhiên, TF không xác định xem từ đó quan trọng trong ngữ cảnh của toàn bộ tập dữ liệu.

- **Inverse Document Frequency (IDF):**
  - IDF đo lường mức độ quan trọng của từng từ trong toàn bộ tập dữ liệu, được  tính toán bằng cách lấy tổng số văn bản trong tập dữ liệu và chia cho số văn bản mà từ đó xuất hiện.
  - IDF cho biết khả năng phân biệt của từ trong toàn bộ tập dữ liệu. Các từ xuất hiện trong nhiều tài liệu được coi là không quan trọng trong việc phân biệt nội dung của một tài liệu cụ thể, trong khi các từ xuất hiện trong ít tài liệu được coi là quan trọng hơn.Tóm lại, từ xuất hiện ít trong nhiều văn bản sẽ có IDF cao và ngược lại.

- **Tại sao cần sử dụng cả TF và IDF cùng nhau:** Khi sử dụng cả TF và IDF cùng nhau, ta có Term Frequency-Inverse Document Frequency (TF-IDF),  điều này giúp tạo ra một công cụ mạnh mẽ để hiểu, phân loại, tìm kiếm và trích xuất thông tin từ các tài liệu văn bản, giúp cải thiện hiệu suất và độ chính xác trong xử lý ngôn ngữ tự nhiên.
  - Tích hợp thông tin ngữ cảnh và tần số: Khi chỉ sử dụng TF  mà không tính tới IDF, có thể bị thiếu sót bởi những từ xuất hiện nhiều lần trong tất cả văn bản, nhưng không phải lúc nào cũng quan trọng. IDF giúp bạn phát hiện được sự khác biệt giữa từng từ trong ngữ cảnh của toàn bộ tập dữ liệu.
  - Xác định tính độc đáo: TF-IDF giúp xác định các từ có tính độc đáo, xuất hiện ít trong một số văn bản nhưng lại xuất hiện trong một văn bản cụ thể. Điều này giúp xác định các từ khóa quan trọng trong văn bản.
  - Đánh giá mức độ quan trọng: TF-IDF đánh giá mức độ quan trọng của một từ trong ngữ cảnh của cả văn bản cụ thể và toàn bộ tập dữ liệu. Nó giúp phát hiện sự khác biệt giữa văn bản dựa trên sự xuất hiện của các từ hoặc cụm từ.
  - Tự động điều chỉnh trọng số: TF-IDF cung cấp một cách tự động để đánh giá trọng số của từng từ hoặc cụm từ trong một văn bản cụ thể. Nếu một từ xuất hiện nhiều lần trong văn bản nhưng xuất hiện ở ít văn bản khác, nó sẽ có trọng số cao hơn, chỉ ra tính độc đáo và quan trọng của nó.
  - Lọc bỏ stopwords: Các từ thông thường như "và," "là," "trong" thường có TF cao trong tài liệu nhưng ít có ý nghĩa đối với nội dung thực sự. Nhờ vào IDF, những từ này có thể có giá trị IDF thấp, giúp làm giảm tầm quan trọng của chúng trong hệ thống TF-IDF. Điều này giúp tập trung vào các từ có tính nội dung quan trọng hơn và loại bỏ nhiễu không cần thiết.
  - Tìm ra các từ hoặc cụm từ đặc trưng định danh nội dung của văn bản. Khi kết hợp cả TF và IDF, các từ mà xuất hiện thường xuyên trong tài liệu cụ thể và ít xuất hiện trong tập hợp tài liệu khác sẽ có giá trị TF-IDF cao. Do đó, các từ với giá trị TF-IDF cao thường đại diện cho các thuộc tính quan trọng của tài liệu và có khả năng định danh nội dung.
  - Hiệu suất tìm kiếm tốt hơn:  sử dụng TF-IDF trong các công cụ tìm kiếm, có thể nhanh chóng tìm thấy các văn bản liên quan dựa trên cả nội dung và tần số xuất hiện của các từ. Điều này giúp tối ưu hóa hiệu suất tìm kiếm thông tin.
  - Phân loại và gom nhóm: TF-IDF cung cấp cơ sở tốt để phân loại văn bản vào các danh mục hoặc gom nhóm chúng lại với nhau dựa trên đặc điểm nội dung và sự tương tự giữa các văn bản.
  

## Câu 3
Evaluate the advantages and disadvantages of using TF-IDF in text processing.

**Ưu điểm của TF-IDF:**

- Đo lường tầm quan trọng của từng từ: TF-IDF giúp xác định mức độ quan trọng của từng từ trong văn bản. Các từ có giá trị TF-IDF cao thường đại diện cho nội dung đặc biệt hoặc quan trọng của tài liệu.
- Loại bỏ stop words: TF-IDF có thể loại bỏ các từ ngữ không quan trọng (stop words) một cách tự động, giúp tăng hiệu quả trong việc xác định sự khác biệt giữa các tài liệu.
- Đa dạng hóa văn bản: TF-IDF có thể được sử dụng để tạo vector biểu diễn văn bản, giúp tìm hiểu mối quan hệ giữa các tài liệu và tìm kiếm tài liệu tương tự.
- Phân biệt tài liệu: TF-IDF giúp phân biệt các tài liệu dựa trên sự khác biệt về nội dung. Các từ quan trọng được làm nổi bật trong việc xác định tính khác biệt giữa các tài liệu.
- Phù hợp với tìm kiếm thông tin: TF-IDF thường được sử dụng trong các hệ thống tìm kiếm thông tin, giúp xác định xem một tài liệu có liên quan đến một truy vấn cụ thể hay không.
- Không cần giám sát: TF-IDF không đòi hỏi sự giám sát, nghĩa là không cần nhãn dữ liệu hoặc thông tin thêm để làm việc. Điều này làm cho nó phù hợp trong việc tiền xử lý dữ liệu.
- Tính nhất quán và dễ hiểu: TF-IDF là một phương pháp đơn giản và dễ hiểu. Người dùng có thể dễ dàng xác định tại sao một từ có giá trị TF-IDF cao trong tài liệu cụ thể.
- Dễ dàng triển khai: TF-IDF là một phương pháp đơn giản và hiệu quả, có thể triển khai dễ dàng trong nhiều ngôn ngữ lập trình và hệ thống.

**Hạn chế của TF-IDF:**
- Không hiểu nghĩa: TF-IDF không hiểu nghĩa của từng từ hoặc cụm từ, chỉ xem xét tần suất xuất hiện. Điều này đôi khi dẫn đến việc không phân biệt giữa các loại tài liệu có cùng từ khóa. Do đó, nó có thể không hiệu quả trong việc xử lý các ngôn ngữ tự nhiên phức tạp với nghĩa phụ thuộc vào ngữ cảnh.
- Không xử lý ngữ cảnh: TF-IDF không xem xét ngữ cảnh của từng từ trong văn bản. Điều này có nghĩa rằng nó không hiểu mối quan hệ giữa các từ trong cùng một văn bản.
- Nhạy cảm với thay đổi văn bản: TF-IDF có thể thay đổi đáng kể nếu bạn thêm hoặc xóa một từ nhỏ trong văn bản, điều này có thể làm cho việc so sánh tài liệu trở nên không ổn định.
- Không phân biệt từ đồng âm: TF-IDF không phân biệt giữa các từ đồng âm, tức là các từ có cùng cách viết nhưng có nghĩa khác nhau.
- Không phân biệt thứ tự từ: TF-IDF không xem xét thứ tự từ, điều này có nghĩa rằng nó không biết sự xuất hiện của các từ theo thứ tự cụ thể trong văn bản. Một văn bản có cấu trúc từ vựng tốt có thể không được đánh giá cao.

Điều này có nghĩa  những từ thực sự quan trọng có thể bị đánh giá thấp hơn so với các từ phổ biến. Nhưng với số lượng từ giới hạn trong văn bản ngắn, giá trị TF-IDF có thể không phản ánh chính xác tầm quan trọng của các từ đó.
- Phụ thuộc vào việc tiền xử lý dữ liệu: Kết quả của TF-IDF có thể bị ảnh hưởng bởi các bước tiền xử lý dữ liệu như loại bỏ stop words, stemming hoặc lemmatization.

Tóm lại, TF-IDF là một công cụ hữu ích trong xử lý văn bản, nhưng nó có giới hạn trong việc hiểu sâu hơn về nội dung và ngữ cảnh. ## Câu 4
In the scikit-learn library, investigate how IDF is implemented. Compare it to the standard IDF formula and explain any differences if present.

- Trong thư viện scikit-learn, IDF (Inverse Document Frequency) được thực hiện thông qua lớp TfidfVectorizer. Lớp TfidfVectorizer cung cấp sự linh hoạt cho việc tùy chỉnh việc tiền xử lý và trích xuất đặc trưng theo nhu cầu cụ thể. Cách IDF được triển khai trong scikit-learn:
  - Nhập module cần thiết: Đầu tiên, cần nhập lớp **TfidfVectorizer** từ module **"sklearn.feature_extraction.text"**.
  - Tạo một đối tượng của lớp **TfidfVectorizer**.Lớp này cho phép chuyển đổi một tập hợp các tài liệu văn bản thành ma trận đặc trưng TF-IDF.
  - Fit và chuyển đổi dữ liệu của bạn: fit và chuyển đổi tập hợp tài liệu của bạn bằng phương thức **fit_transform** của TfidfVectorizer. Điều này sẽ tính toán các giá trị IDF cho mỗi từ trong tập hợp và tính toán các điểm TF-IDF cho các từ trong mỗi tài liệu.
  - Truy cập các giá trị IDF: truy cập các giá trị IDF cho mỗi từ bằng cách sử dụng thuộc tính **idf_** của đối tượng **TfidfVectorizer**. Thuộc tính **idf_** chứa  các giá trị IDF cho mỗi từ trong từ vựng được học từ các tài liệu.
  
- Công thức IDF trong scikit-learn:
$$ \text{IDF}(t) = \log\left(\frac{N + 1}{df(t) + 1}\right) + 1 $$
- Công thức trên có một số khác biệt so với công thức IDF tiêu chuẩn (Standard IDF Formula), mà công thức tiêu chuẩn được biểu diễn như sau:
$$ \text{IDF}(t) = \log\left(\frac{N }{df(t)}\right)$$
 Trong đó:
  - $t$: là từ cần tính IDF.
  - $N$: là tổng số tài liệu trong tập dữ liệu.
  - $df(t)$: là số tài liệu chứa từ t.
- So  sánh công thức IDF trong scikit-learn với công thức IDF tiêu chuẩn:
  - Giống nhau:
     - Đều dựa trên ý tưởng chia tổng số tài liệu cho tần suất xuất hiện của từ trong tài liệu ($df(t)$) để tính toán mức quan trọng của từng từ.
     - Đều sử dụng hàm logarithm (log) để tăng đáng kể giá trị của IDF khi tần suất xuất hiện của từ (df(t)) giảm.
  - Khác nhau:
     - Smoothing : Công thức IDF trong scikit-learn thêm "+1" vào cả tử số và mẫu số của công thức. Điều này giúp tránh lỗi chia cho 0 khi một từ không xuất hiện trong bất kỳ tài liệu nào đảm bảo rằng tất cả các từ có giá trị IDF xác định, ngay cả khi chúng không xuất hiện trong bất kỳ tài liệu nào trong tập hợp. Điều này có ý nghĩa quan trọng trong các ứng dụng thực tế để tránh lỗi toán học và đảm bảo rằng tất cả các từ đóng góp vào việc tính toán điểm TF-IDF. Công thức tiêu chuẩn không có smoothing và có thể dẫn đến undefined khi $df(t) = 0$.
    - Tùy chọn sử dụng Smoothing: Công thức trong scikit-learn cộng thêm 1 vào kết quả của công thức. Điều này dẫn đến giá trị IDF trong scikit-learn cao hơn một lượng nhỏ so với standard IDF. Thêm 1 giúp tăng đáng kể giá trị IDF cho các từ có tần suất xuất hiện thấp và giúp tạo sự khác biệt lớn hơn giữa các từ.  Có thể điều chỉnh việc và kiểm soát giá trị IDF thông qua tham số smooth_idf. Nếu smooth_idf=True (đây là giá trị mặc định), thì smoothing được áp dụng; nếu smooth_idf=False, thì standard IDF được sử dụng mà không có sự làm mịn bổ sung có nghĩa chấp nhận việc công thức IDF có thể trả về undefined cho các từ không xuất hiện trong bất kỳ tài liệu nào.

## Câu 4
Write a program to calculate the term frequency (TF) of each word in the reviews.

```python
def compute_tf(vocab, sentence):
    """
    Tính toán TF (Term Frequency) của mỗi từ trong các đánh giá.

    Tham số:
    - vocab: Danh sách các từ vựng mà bạn muốn xem xét.
    - sentence: review đã qua xử lý và stem, dưới dạng danh sách các từ.

    Kết quả:
    - tf_dict: Một từ điển với các từ làm khóa và giá trị tương ứng là điểm TF.
    """
    tf_dict = {}
    total_words = len(sentence)

    for word in vocab:
        # Đếm số lần từ xuất hiện trong câu
        word_count = sentence.count(word)
        # Tính toán điểm TF
        tf = word_count / total_words
        tf_dict[word] = tf

    return tf_dict
```

```python
def stem_words(sentence):
    stemmer = nltk.PorterStemmer()
    stemmed_sentence = [stemmer.stem(word) for word in sentence]
    return stemmed_sentence

def create_vocab_counts(df):
    vocab_counts_dict = {}
    for index, row in df.iterrows():
        review = row['preprocessed_review']
        stemmed_review = stem_words(review)
        vocab_counts = {}
        for word in stemmed_review:
            if word in vocab_counts:
                vocab_counts[word] += 1
            else:
                vocab_counts[word] = 1
        vocab_counts_dict[index] = vocab_counts
    vocab_counts_df = pd.DataFrame.from_dict(vocab_counts_dict, orient='index')
    return vocab_counts_df

vocab_counts_df = create_vocab_counts(df.copy())
vocab_counts_df = vocab_counts_df.fillna(0)
vocab_counts_df
```

```python
# Get 5000 words common
common_words_5000 = vocab_counts_df.sum(axis=0).sort_values(ascending=False)[:5000].index
vocab_counts_df[common_words_5000]
```

```python
# Ví dụ sử dụng hàm compute_tf cho đánh giá đầu tiên trong DataFrame
review_index = 0  # Thay đổi chỉ số này để tính TF cho đánh giá cụ thể
sentence = df['preprocessed_review'][review_index]
tf_scores = compute_tf(common_words_5000, sentence)
```

## Câu 5
Write a program to calculate the inverse document frequency (IDF) of words in the reviews, considering each review as a separate document.

```python
import math

def computeIDF(preprocessed_reviews, word_list):
    """
    Tính toán IDF (Inverse Document Frequency) của các từ trong các đánh giá.

    Tham số:
    - preprocessed_reviews: Một danh sách các đánh giá đã qua xử lý và stem (mỗi đánh giá là một danh sách các từ).
    - word_list: Danh sách các từ mà bạn muốn tính IDF.

    Kết quả:
    - idf_dict: Một từ điển với các từ làm khóa và giá trị tương ứng là điểm IDF của chúng.
    """
    idf_dict = {}
    total_reviews = len(preprocessed_reviews)

    for word in word_list:
        # Đếm số lần từ xuất hiện trong các đánh giá
        reviews_with_word = sum(1 for review in preprocessed_reviews if word in review)
        # Tính toán điểm IDF
        idf = math.log(total_reviews / (reviews_with_word + 1))  # Thêm 1 để tránh chia cho 0
        idf_dict[word] = idf

    return idf_dict
```

```python
# Ví dụ sử dụng hàm computeIDF với các đánh giá đã qua xử lý và danh sách top_5000_words
idf_scores = computeIDF(df['preprocessed_review'], common_words_5000)
```

## Câu 6
Write a function that takes a review sentence as input and returns its TF-IDF representation using the information obtained in the previous steps.

```python
def compute_tf_idf(idf_dict, tf_vector):
    tfidf_vector = {}

    # Tính điểm TF-IDF cho từng từ
    for word in idf_dict:
        tfidf_vector[word] = tf_vector[word] * idf_dict[word]

    return tfidf_vector

```
